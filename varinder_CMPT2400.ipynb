## **Understanding The Data**


### **Dataset Overview**
Our group is conducting an exploratory data analysis (EDA) using the National Pollutant Release Inventory (NPRI) dataset, provided by Environment and Climate Change Canada (ECCC). This comprehensive dataset covers the reporting period from 2000 to 2022, segmented into three primary tables: releases, disposals and transfers, and comments. These tables capture a wide range of pollutant data reported annually by Canadian facilities. The dataset is structured to monitor pollutants across various media—air, water, and land—offering crucial insights into industrial environmental impacts and trends over time.

### **Key Features of the Dataset**
The NPRI dataset comprises several essential features that are critical for understanding pollutant behavior, facility characteristics, and geographical information. Some of the key attributes include:

1. *Reporting Year:* Tracks the year in which the data was reported.
2. *NPRI_ID:* A unique identifier for each facility reporting to the NPRI.
3. *Number of Employees:* A proxy for facility size, potentially impacting pollutant levels.
4. *Facility Name, NAICS Code, Province, City:* Identifies the reporting entity and geographical information, allowing for regional analysis of pollution trends.
5. *Geospatial Data (Latitude, Longitude):* Enables spatial analysis, such as mapping pollutant sources and identifying hotspots.
6. *Substance Name and CAS Number:* Specifies the type of pollutant, from common air contaminants (e.g., carbon monoxide) to more toxic chemicals (e.g., heavy metals, persistent organic pollutants).
7. *Units and Estimates:* Standardizes how pollutant quantities are measured and estimated.
8. *Releases to Air, Water, and Land:* Detailed breakdown of releases by medium, further divided into subcategories such as fugitive air emissions, stack/point air emissions, and spills into water bodies.
9. *Total Releases to All Media (<1 Tonne):* Provides a summed total for smaller release quantities, which, when aggregated, can still significantly affect ecosystems.<br>
The diversity of this dataset enables multi-dimensional analyses of pollutant trends, regulatory compliance, and environmental risks associated with specific industries.

### **Advanced Research and Insights on the NPRI Dataset**
The NPRI serves as an integral part of Canada's national environmental strategy. It acts as a public database that tracks over 320 pollutants across thousands of facilities nationwide, giving stakeholders—such as government agencies, environmental researchers, and policy-makers—access to critical data for informed decision-making. The dataset’s richness facilitates the analysis of both direct emissions (e.g., stack emissions, leaks, spills) and indirect transfers (e.g., waste treatment, recycling), allowing for a holistic view of how industrial activities affect the environment.

### **Environmental Impacts of Pollutants**
The substances tracked in this dataset pose varying levels of risk to human health and the ecosystem. Many of the pollutants are criteria air contaminants (CACs), which contribute to issues such as smog, acid rain, and respiratory illnesses. For instance, pollutants like nitrogen oxides (NOx) and volatile organic compounds (VOCs) are key precursors to ground-level ozone, a significant component of urban smog. Chronic exposure to such pollutants has been linked to severe health conditions, including cardiovascular and respiratory diseases.

In addition to air pollution, the dataset provides insight into the disposal and release of hazardous substances into soil and water. Facilities that dispose of waste in landfills, or that engage in improper waste handling, create long-term environmental risks. Pollutants like heavy metals (e.g., mercury, cadmium) and persistent organic pollutants (POPs) can accumulate in water bodies, threatening aquatic ecosystems and entering the food chain. The NPRI data allow for the identification of regions that may face these environmental stressors, facilitating targeted remediation efforts.

### **Pollutant Management Strategies**
An important feature of the NPRI dataset is its tracking of how pollutants are managed. Facilities are required to report not just releases, but also how pollutants are treated or recycled. This helps assess the effectiveness of industrial pollution control technologies and waste management practices. For example, comparing on-site treatment rates to off-site disposal can reveal disparities in environmental responsibility between facilities.

### **Regional and Temporal Trends**
By leveraging the geospatial and temporal dimensions of the dataset, we can analyze how pollution trends evolve across regions and over time. This allows us to investigate:

1. *Industrial hotspots:* Regions where pollution is consistently high, which may necessitate stronger regulations or enforcement actions.
2. *Temporal patterns:* Identifying whether regulatory policies, such as the Canadian Environmental Protection Act (CEPA), have effectively curbed emissions over the past two decades.
3. *Cross-industry comparisons:* Analyzing different NAICS categories can reveal which industries contribute the most to specific pollutants and which have improved their environmental performance.
### **Predictive Analysis of Environmental Risk**
For our project, we aim to explore not only past trends but also use the dataset to predict future pollution levels, particularly focusing on criteria air contaminants like sulphur dioxide (SO₂), nitrogen oxides (NOx), particulate matter (PM), and carbon monoxide (CO). By applying advanced machine learning models to this dataset, we can identify potential future risks, allowing for proactive environmental management. These predictions will be invaluable for government agencies seeking to meet sustainability goals and for communities concerned about the long-term impacts of industrial activities on public health and the environment.

**Issue:** The dataset contains multiple units for pollutant releases: tonnes, kilograms (kg), grams, and g TEQ.

**Solution:** We can convert all the measurements to single unit, usually for larger values "tonnes" unit is used. We already got some selected columns for releases and we will perform this function on those columns.<br>
Convert the values of "Total_releases" column to "Tonnes". This code converts the values in the Total_Releases column of the df_releases DataFrame to tonnes based on the units specified in the Units / Unités column. The function "convert_to_tonnes()" handles the conversions from kilograms, grams, and grams TEQ to tonnes. The ".apply()" method applies this function to each row, ensuring consistent measurement units across the dataset. Standardizing these units facilitates accurate analysis and comparison of pollutant releases.

```python
# Step 1: Define a function to convert values to tonnes
def convert_to_tonnes(value, unit):
    if unit == 'kg':
        return value * 0.001  # Convert kg to tonnes
    elif unit == 'grams':
        return value * 0.000001  # Convert grams to tonnes
    elif unit == 'g TEQ':
        return value * 0.000001  # Convert g TEQ to tonnes
    else:  # If already in tonnes or unrecognized unit, return the value as-is
        return value
# Step 2: Apply the conversion directly to the 'Total_Releases' column
df_releases['Total_Releases'] = df_releases.apply(
    lambda row: convert_to_tonnes(row['Total_Releases'], row['Units / Unités']), axis=1
)

# Display the updated DataFrame with the standardized 'Total_Releases' in tonnes
print(df_releases[['Total_Releases', 'Units / Unités']].head())
```
